```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("C:/Users/Muhammad/Houseprices.xlsx")  
df.head() 
```

```python


plt.xlabel("Area")
plt.ylabel("Price")
plt.scatter(df["area"], df["price"], color="red", marker="+")
plt.show()
```

```python
new_df=df.drop("price",axis='columns')
new_df
```

```python
model=linear_model.LinearRegression()
model.fit(df[['area', 'bedroom', 'age']].values,df['price'].values)
```

```python
model.predict([[5000,4,5]])
```

```python
import pickle
```

```python
with open('model_pickle','wb') as file:
    pickle.dump(model,file)

```

```python
with open('model_pickle','rb') as f:
    mp= pickle.load(f)
```
